```python
%sql
CREATE SCHEMA IF NOT EXISTS retail_prod.qa;
```

```python
%sql
drop table if exists retail_prod.qa.run_summary;
drop table if exists retail_prod.qa.run_detail_summary;
```

```python
%sql
CREATE TABLE retail_prod.qa.run_summary (
    Product_Name STRING,
    Testing_Layer STRING,
    No_of_Test_Cases_Errored INT,
    No_of_Test_Cases_Failed INT,
    No_of_Test_Cases_Passed INT,
    Protocol_File_Path STRING,
    Test_Environment STRING,
    Test_Protocol_End_Time STRING,
    Test_Protocol_Name STRING,
    Test_Protocol_Start_Time STRING,
    Test_Protocol_Version STRING,
    Testcase_Type STRING,
    Testcases_Executed STRING,
    Total_No_of_Test_Cases INT,
    Total_Protocol_Run_Time STRING,
    Test_Protocol_Run_ID STRING,
    Test_Protocol_Run_Name STRING,
    Check_Name STRING
)
USING DELTA;
```

```python
%sql
CREATE TABLE retail_prod.qa.run_detail_summary (
    Test_Protocol_Run_ID STRING,
    Test_Protocol_Run_Name STRING,
    Check_Name STRING,
    Testcase_Name STRING,
    No_of_Rows_in_Source BIGINT,
    No_of_Rows_in_Target BIGINT,
    Distinct_Rows_Matched BIGINT,
    Distinct_Rows_Mismatched BIGINT,
    No_of_Distinct_Rows_in_Source BIGINT,
    No_of_Distinct_Rows_in_Target BIGINT,
    No_of_Records BIGINT,
    Total_No_of_Checks INT,
    No_of_Checks_Passed INT,
    No_of_Checks_Failed INT,
    Test_Result STRING,
    Reason STRING,
    Runtime STRING
)
USING DELTA;
```
